```python

dates = []
for year in range(start_year,end_year):
    for quarter in ['03','06','09','12']:
        temp = str(year)+'-'+quarter
        dates.append(temp)
        
df = pd.DataFrame()
## Grab the files
for each in dates:
    try:
        try:
            url = start +each + '.zip'
            ## Using Zipfiles, we can extract the files of interest from the zip files
            resp = urlopen(url)
            print(url)
            zipfile = ZipFile(BytesIO(resp.read()))
        except:
            try:
                url = alt_url +each.split('-')[0]+each.split('-')[1] + '.zip'
                ## Using Zipfiles, we can extract the files of interest from the zip files
                resp = urlopen(url)
                print(url)
                zipfile = ZipFile(BytesIO(resp.read()))
            except:
                try:
                    url = alt_url +each.split('-')[0]+each.split('-')[1] + '.Zip'
                    ## Using Zipfiles, we can extract the files of interest from the zip files
                    resp = urlopen(url)
                    print(url)
                    zipfile = ZipFile(BytesIO(resp.read()))
                except:
                    ## In 2013-06 they completely changed the naming convention just for the one time. 
                    ## This sectiomn catches that odd one
                    if each == '2013-06':
                        url = 'https://www.ncua.gov/files/publications/data-apps/5300Data0613Final.zip'
                    else:
                        url = 'https://www.ncua.gov/files/publications/analysis/Call-Report-Data-'+each+'.zip'
                    ## Using Zipfiles, we can extract the files of interest from the zip files
                    resp = urlopen(url)
                    print(url)
                    zipfile = ZipFile(BytesIO(resp.read()))

        ## Extract the file with the descriptions of the heads(spelling errors are common so we try and except)
        try:
            ziped = zipfile.open('AcctDesc.txt')
        except:
            try:
                ziped = zipfile.open('Acct_Desc.txt')
            except:
                ziped = zipfile.open('Acct_Des.txt')
        acct_desc_df = pd.read_csv(ziped, encoding = 'ISO-8859-1')

        ## Extract the file with the metrics 
        try:
            ziped = zipfile.open(file)
        except:
            ziped = zipfile.open(file_upper)
        fs220_df = pd.read_csv(ziped, encoding = 'ISO-8859-1')
        print(each, fs220_df.shape)
        ## Create empty dataframe for us to store the columnn name keys in and transform them to lower case
        column_names = pd.DataFrame()
        ## Store the keys in lower case
        column_names['fs220'] = fs220_df.columns.str.lower()

        ## store the other keys in lowercase
        acct_desc_df['Account'] = acct_desc_df['Account'].str.lower()

        ## Left join the column names that we need to replace from keys to description with the file that has the tranlated values
        new_columns = pd.merge(column_names, acct_desc_df, how = 'left', left_on = 'fs220', right_on = 'Account' )

        ## Some Values were nulls so we replace them with their original code value.
        new_columns['AcctName'].fillna(new_columns['fs220'],inplace = True)

        ## Replace the original dataset column names with the more descriptive titles. 
        fs220_df.columns = fs220_df.columns +" - " + new_columns['AcctName'] 

        df = pd.concat([df,fs220_df],axis =0, sort =False)
    except:
        print("No data for - ", each)
df.to_csv('NCUA_Call_Report.csv', index = False)
```
